```python
import cirq
import numpy as np

def qrng(bit_length: int) -> str:
    # qubits equal to the bit length
    qubits = [cirq.GridQubit(0, i) for i in range(bit_length)]

    #circuit
    circuit = cirq.Circuit()
    for qubit in qubits:
        circuit.append(cirq.H(qubit))
        circuit.append(cirq.measure(qubit))
    simulator = cirq.Simulator()
    result = simulator.run(circuit, repetitions=1)
    random_bits = ''.join(str(result.measurements[str(qubit)][0][0]) for qubit in qubits)

    return random_bits

def gotp(length: int) -> str:

    binary_otp = qrng(length * 3)
    otp_number = str(int(binary_otp, 2))[:length]  # Convert binary to decimal
    return otp_number
n = int(input("Enter the length of OTP: "))
otp = gotp(n)
print("Quantum OTP:", otp)
```
